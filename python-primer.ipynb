# Section 3: Functions

## 3.1 Defining Functions

Functions are reusable blocks of code. Like creating a protocol you can repeat.

````python
def calculate_dice_score(prediction, ground_truth):
    """
    Calculate Dice coefficient (overlap measure).

    Parameters:
    - prediction: model's segmentation
    - ground_truth: expert annotation

    Returns:
    - dice: overlap score (0 to 1, higher is better)
    """
    intersection = (prediction * ground_truth).sum()
    dice = (2 * intersection) / (prediction.sum() + ground_truth.sum())
    return dice

# Use the function
pred = [1, 1, 0, 1, 0]
truth = [1, 1, 1, 1, 0]
score = calculate_dice_score(pred, truth)
print(f"Dice score: {score:.2f}")
````

## 3.2 Function with Default Parameters

````python
def print_results(dice, threshold=0.8):
    """
    Print results with optional threshold.
    """
    print(f"Dice score: {dice:.3f}")
    if dice >= threshold:
        print("✓ Meets quality threshold")
    else:
        print("✗ Below quality threshold")

# Use with default threshold
print_results(0.85)

print("\n---\n")

# Use with custom threshold
print_results(0.85, threshold=0.9)
````

## 4.2 List Operations

````python
# Get length
files = ["img1.nii.gz", "img2.nii.gz", "img3.nii.gz"]
print(f"Total files: {len(files)}")

# Slicing (getting parts of a list)
first_two = files[:2]      # First 2 items
last_one = files[-1]       # Last item
middle = files[1:3]        # Items at index 1 and 2

print(f"First two: {first_two}")
print(f"Last one: {last_one}")
print(f"Middle: {middle}")
````

## 5.2 NumPy: Working with Arrays

NumPy is essential for numerical operations. Medical images are NumPy arrays!

````python
import numpy as np

# Create arrays
pixels = np.array([100, 150, 200, 180, 120])

# Array operations
print(f"Shape: {pixels.shape}")           # Dimensions
print(f"Mean intensity: {pixels.mean()}") # Average
print(f"Max intensity: {pixels.max()}")   # Maximum
print(f"Min intensity: {pixels.min()}")   # Minimum

# Element-wise operations
normalized = pixels / 255.0  # Normalize to 0-1
print(f"Normalized: {normalized}")
````
